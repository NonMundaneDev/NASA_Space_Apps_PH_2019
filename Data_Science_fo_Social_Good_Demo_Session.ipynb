## Problem

[Kiva.org](https://www.kiva.org) is an online crowdfunding platform to extend financial services to poor and financially excluded people around the world, and of course that includes Nigeria. We want to analyze the poverty levels of people in the regions that funds were loaned so we can estimate the poverty levels in these regions, and hopefully help Kiva.org disburse loans more efficiently. <br/>


---


## Objective
To estimate the poverty levels of residents in the regions where Kiva has active loans, and more specifically Nigeria. And also to estimate the counties, sectors and regions (in Nigeria)  that borrow funds the most from this platform.



---



## What We Will Be Analysing?
1. The number of loans taken from the platform (in 2 years) in each country of lenders' residence.

2. The sectors peoeple chose to take out loans for; maybe for education, or entertainment, or food reasons, and so on.

3. We will also be looking at the number of loans taken out by their duration; whether they are long-term or short-term loans.

4. We will also be analysing the number of loans taken out based on gender.

## Why Will this Be Useful?
1. Analyze (at a small scale) which of the countries take out loans the most, and are most likely in need.
2. Examine which of the regions in Nigeria take out loans the most.
3. What the loans are mostly used for to enable the Government to take data-driven actions

```
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
import pylab
import seaborn as sns
plt.style.use('fivethirtyeight')
%matplotlib inline
pylab.rcParams['figure.figsize'] = (10.0, 8.0)
```

### Load your data into the project using the Pandas tool.
loans_df reps general data from Kiva loans
<br/>


```
loans_df = pd.read_csv("kiva_loans.csv", parse_dates=['disbursed_time', 'funded_time', 'posted_time'])
```

### Read the first 5 rows of your data (just like you would do, in an Excel table).

```
loans_df.head()
```

### Visualize the number of loans taken by countries on a graph.

```
pylab.rcParams['figure.figsize'] = (8.0, 25.0)
plt.style.use('fivethirtyeight')
loans_df.groupby(loans_df.country).id.count().sort_values().plot.barh(color='cornflowerblue');
plt.ylabel('Loan Count')
plt.title("Loan Count by Country");
```

### Visualize the number of loans based on categories/sectors the loans were taken out for.

```
pylab.rcParams['figure.figsize'] = (6.0, 6.0)
loans_df.groupby(loans_df.sector).id.count().sort_values().plot.bar(color='cornflowerblue');
plt.ylabel('Loan Count')
plt.title("Loan Count by Sector");
```

### Visualize the number of short and long-term loans taken out.

```
loans_df.term_in_months.plot.hist(bins=100);
plt.ylabel('Loan Count')
plt.title("Loan Count by Loan Duration");
```

### Visualize the number of loans taken by gender.

```
## Define a function to help seperate each genders and classify males as 'M'...
## ... and females as 'F' or mixed groups (male and female) as 'MF'

def process_gender(x):
    
    if type(x) is float and np.isnan(x):
        return "nan"
    genders = x.split(",")
    male_count = sum(g.strip() == 'male' for g in genders)
    female_count = sum(g.strip() == 'female' for g in genders)
    
    if(male_count > 0 and female_count > 0):
        return "MF"
    elif(female_count > 0):
        return "F"
    elif (male_count > 0):
        return "M"
```

```
## Call the function you just created and apply it to the "borrower_genders" column from our dataset
loans_df.borrower_genders = loans_df.borrower_genders.apply(process_gender)
```

### Graphically represent the number of loans borrowed by each gender.

```
## Plot the a bar chart of the data for a graphical representation.
loans_df.borrower_genders.value_counts().plot.bar(color='cornflowerblue');
plt.xlabel('Borrower Group/Individual Gender')
plt.ylabel('Count')
plt.title("Loan Count by Gender of Borrower");
```

```
## set Pandas to display full rows.

pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)

# Check the first 5 rows of data for info on Nigeria

loans_df[loans_df.country == 'Nigeria'].head()
```

```
## Create a dataframe for Nigeria from the data

df_nigeria = loans_df[loans_df.country == 'Nigeria']
```

```
# Check the  number of borrowers from Nigeria.

df_nigeria.shape

# The first value (not 20) reps the number of borrowers from Nigeria.
```

```
# Visualize the number of loans taken by gender in Nigeria.

df_nigeria.borrower_genders.value_counts().plot.bar(color='cornflowerblue');
plt.xlabel('Borrower Group/Individual Gender')
plt.ylabel('Count')
plt.title("Loan Count by Gender of Borrower in Nigeria");
```

```
# Visualize the number of loans taken out for each category by borrowers from Nigeria

pylab.rcParams['figure.figsize'] = (6.0, 6.0)
loans_df[loans_df.country == 'Nigeria'].sector.value_counts().plot.bar(color='cornflowerblue');
plt.title("Loan Count by Sector in Nigeria")
plt.xlabel("Sector")
plt.ylabel("Loan Count")
```

```
# Visualize the number of loans taken by region.

pylab.rcParams['figure.figsize'] = (8.0, 8.0)
df_nigeria.groupby('region').id.count().sort_values(ascending=False).head(20).plot.bar(color='cornflowerblue');
```
